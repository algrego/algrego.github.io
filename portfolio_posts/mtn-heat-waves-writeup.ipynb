# **Mountain Town Heat Waves:** 
## An Exploration of Socioeconomic Characteristics and Climate Impacts of Three Rocky Mountain Ski Town Regions

### **Introduction and Significance**
According to the [United Nations]( https://www.un.org/en/climatechange/climate-adaptation), nearly half of the global population is currently highly vulnerable to climate change impacts. This includes droughts, flooding, storms, heat stress, and food insecurity. The US is not exempt to experience the extreme impacts and weather events brought on by climate change. 

There have been many attempts to quantify the risks associated with climate change and socioeconomic differences. But very few tools exist to pull all these research together. Notably, the U.S. Climate Vulnerability Index lists Aspen/Vail (Colorado), Teton County (Wyoming), and Gallatin County (Montana) as the least vulnerable regions to climate change in the U.S. This index is the closest representation of the research we are conducting, so future research will look into ways we can improve this existing index.

Mountain towns are particularly interesting to research in this context as their economies are dependent on weather related recreation and tourism, they often have high socioeconomic inequities, and can experience extreme events such as flooding, droughts, and heat stress. It is also a unique area of study because there are also many vacation or second homes which provides a unique perspective. 

This research can be used by federal, state, and city planners and policy makers to better understand where resources should be targeted within their region. Areas with high social vulnerability and high climate impacts may need significantly more targeted resources to support their community. There is [“growing recognition that climate change disproportionately affects marginalized communities – those least responsible for greenhouse gas emissions and historically disadvantaged through dispossession, system racism, and exclusionary development”] (https://www.sciencedirect.com/science/article/pii/S0959378025000640). This work can help make targeted steps toward quantifying these inequities.


### **Background**

For this case study we will be evaluating three Rocky Mountain Ski Town regions: Aspen/Vail (Colorado), Jackson Hole (Wyoming), and Big Sky (Montana). These regions have similar sizes and tourism economies, but they are in different states which allows us to evaluate different types of climate policies that occur in each of these areas. For example, Colorado has a [tool](https://dlg.colorado.gov/measuring-climate-and-social-risk-and-vulnerabilities) to assess Climate and Social Vulnerabilities so the state can better plan to support communities moving forward.

### **Data Sources**

The socioeconomic data for this project is taken from the [CDC’s Social Vulnerability Index](). Our current assessment used county tracts, but future work may pivot to census tract data to get a higher resolution perspective. One of the variables of focus is the percentage of residents who are experiencing housing cost burden. This means that the over 30% of a home’s income is spent on housing costs. Other CDC socioeconomic variables that may be relevant in climate adaptation strategies include percent poverty, crowding, and uninsured. The [CDC’s social vulnerability index (SVI)]( https://svi.cdc.gov/map25/data/docs/SVI2022Documentation_ZCTA.pdf) includes over 10 variables to interpret, so ideally we will create a fuzzy logic model using the most relevant variables to create a climate adaptability score.  Future research could also look into other publicly available data sets such [Resilience Analysis and Planning Tool (RAPT) by FEMA]( https://www.fema.gov/emergency-managers/practitioners/resilience-analysis-and-planning-tool) which looks at variables such as tornados, wildfires, and severe weather. This data also includes additional relevant variables such as tribal boundaries and locations of important infrastructure like schools and hospitals.

For the climate analysis, we are starting with Maximum Air Temperature (tasmax) from the MACAv2 dataset. To have the most accurate climate data, we selected four models scenarios (Warm-Wet, Warm-Dry, Cold-Wet, Cold-Dry) and corresponding models using the RCP4.5 pathway. We looked at historical data from 1970-1999 and future projections of 2036-2065. Future work will attempt to combine these models to see the best results and then compare what regions will experience the most dramatic maximum temperatures.

### **Summary / Conclusions**
The socioeconomic analysis revealed that different variables across households such as housing cost burden and poverty look different across the regions and could be studied from a geodatabase perspective. This is meaningful because it allows us to get a full picture of how these areas may experience extreme climate events.

The initial climate analysis also revealed that there are distinct differences in the models used for climate projects and a combined version of models can be the most effective to understand future patterns.
